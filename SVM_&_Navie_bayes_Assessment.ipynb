1- What is a Support Vector Machine (SVM)?
Ans- A Support Vector Machine (SVM) is a machine learning method used to classify data into different groups. It works by finding the best line or boundary that separates the data points of different categories. The main goal is to create a gap (called a margin) between the groups that is as wide as possible, so new data can be classified accurately. The closest points to this line are called support vectors, and they help decide where the line should be. SVMs are useful for things like image recognition, email spam detection, and medical diagnosis.

2- What is the difference between Hard Margin and Soft Margin SVM?
Ans- Hard Margin - no errors allowed, works only for clean data.
Soft Margin - allows small errors, works better for real-world messy data.

3- What is the mathematical intuition behind SVM?
Ans- SVM finds a line w⋅x+b=0 that separates classes while maximizing the margin. It minimizes 1/2 ||w||^2 to get the widest gap between classes.

4- What is the role of Lagrange Multipliers in SVM?
Ans- They help solve SVM's optimization problem by turning constraints into a form that can be efficiently optimized to find the best separating boundary.

5- What are Support Vectors in SVM?
Ans- Support vectors are the data points closest to the decision boundary that help define the position and direction of the separating line.

6- What is a Support Vector Classifier (SVC)?
Ans- A Support Vector Classifier (SVC) is the practical implementation of SVM used to classify data by finding the best separating boundary between classes.

7- What is a Support Vector Regressor (SVR)?
Ans- A Support Vector Regressor (SVR) is an SVM-based model used for regression tasks that fits the best line or curve within a certain error margin.

8- What is the Kernel Trick in SVM?
Ans- The kernel trick allows SVM to handle non-linear data by mapping it into a higher-dimensional space where it becomes linearly separable.

9- Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.
Ans- Linear Kernel: Works well for linearly separable data.
Polynomial Kernel: Captures curved relations using polynomial functions.
RBF Kernel: Handles complex, non-linear patterns by mapping data to high dimensions.

10- What is the effect of the C parameter in SVM?
Ans- The C parameter controls the trade-off between a wide margin and classification errors — a large C means fewer errors but smaller margin, while a small C allows more errors but a wider margin.

11- What is the role of the Gamma parameter in RBF Kernel SVM?
Ans- The Gamma parameter controls how far the influence of a single data point reaches — high gamma means close influence (complex model), low gamma means far influence (simpler model).

12- What is the Naïve Bayes classifier, and why is it called "Naive?
Ans- Naive Bayes is a probabilistic classifier based on Bayes theorem that predicts classes using feature probabilities. It's called naïve because it assumes all features are independent, which is rarely true in real data.

13- What is Bayes' Theorem?
Ans- Bayes' Theorem calculates the probability of an event based on prior knowledge: P(A|B)=[P(B|A)*P(A)]/P(B)
It updates beliefs after seeing new evidence.

14- Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?
Ans- Gaussian NB- Used for continuous data (assumes data follows a normal distribution).
Multinomial NB- Used for count data like word frequencies in text.
Bernoulli NB- Used for binary data (features with 0 or 1 values).

15- When should you use Gaussian Naïve Bayes over other variants?
Ans- We use Gaussian Naive Bayes when features are continuous and roughly follow a normal (bell-shaped) distribution.

16- What are the key assumptions made by Naïve Bayes?
Ans-:1-All features are independent of each other.
2-Each feature equally contributes to the outcome.
3-Data features are conditionally independent given the class label.

17- What are the advantages and disadvantages of Naïve Bayes?
Ans- Advantages: Simple, fast, works well with small data, and good for text classification.
Disadvantages: Assumes feature independence and can perform poorly if features are highly correlated.

18- Why is Naïve Bayes a good choice for text classification?
Ans- Because it handles high-dimensional data well, works efficiently with word frequencies, and performs well even with small training data.

19- Compare SVM and Naïve Bayes for classification tasks.
Ans- SVM: Works well with complex, high-dimensional data; slower to train; focuses on boundaries.
Naive Bayes: Fast, simple, good for text data; assumes feature independence; less accurate on complex data.

20- How does Laplace Smoothing help in Naive Bayes?
Ans- Laplace smoothing prevents zero probabilities by adding a small value (usually 1) to all counts, ensuring every possible outcome has a nonzero probability.

24- Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
# load dataset (2 features for easy plotting)
X, y = datasets.make_classification(n_features=2, n_redundant=0, n_informative=2,
n_clusters_per_class=1, random_state=0)
# train SVM with polynomial kernel
model = SVC(kernel='poly', degree=3)
model.fit(X, y)
# create grid to plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
# predict over grid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
# plot decision boundary and data points
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.title("SVM with Polynomial Kernel")
plt.show()
```

27- Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# create sample data

X, y = datasets.make_classification(n_features=2, n_redundant=0, n_informative=2,
n_clusters_per_class=1, random_state=1)

# try different C values

C_values = [0.1, 1, 10]

# plot decision boundaries for each C

plt.figure(figsize=(10, 3))
for i, C in enumerate(C_values):
    model = SVC(kernel='linear',C=C)
    model.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(1, 3, i + 1)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title("C = " + str(C))


plt.tight_layout()
plt.show()
```

31-Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,
gamma, kernel).

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# load dataset
X, y = datasets.load_iris(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# define parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['linear', 'rbf', 'poly']
}

# create and run grid search
grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(X_train, y_train)

# best model
best_model = grid.best_estimator_

# predict and check accuracy
y_pred = best_model.predict(X_test)
print("Best Parameters:", grid.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

32- Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and
check it improve accuracy.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# create imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5,
                           weights=[0.9, 0.1], random_state=0)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train without class weights
model_unweighted = SVC(kernel='linear')
model_unweighted.fit(X_train, y_train)
y_pred_unweighted = model_unweighted.predict(X_test)

# train with class weights
model_weighted = SVC(kernel='linear', class_weight='balanced')
model_weighted.fit(X_train, y_train)
y_pred_weighted = model_weighted.predict(X_test)

# compare accuracies
print("Accuracy without class weighting:", accuracy_score(y_test, y_pred_unweighted))
print("Accuracy with class weighting:", accuracy_score(y_test, y_pred_weighted))
```

35- Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare
results.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif

# load dataset
X, y = datasets.load_iris(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train Naive Bayes without feature selection
model_full = GaussianNB()
model_full.fit(X_train, y_train)
y_pred_full = model_full.predict(X_test)

# apply feature selection (select best 2 features)
selector = SelectKBest(score_func=f_classif, k=2)
X_train_new = selector.fit_transform(X_train, y_train)
X_test_new = selector.transform(X_test)

# train Naive Bayes with selected features
model_selected = GaussianNB()
model_selected.fit(X_train_new, y_train)
y_pred_selected = model_selected.predict(X_test_new)

# compare accuracies
print("Accuracy without feature selection:", accuracy_score(y_test, y_pred_full))
print("Accuracy with feature selection:", accuracy_score(y_test, y_pred_selected))
```

36- Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO)
strategies on the Wine dataset and compare their accuracy?

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score

# load wine dataset
X, y = datasets.load_wine(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# One-vs-Rest strategy
ovr_model = OneVsRestClassifier(SVC(kernel='linear'))
ovr_model.fit(X_train, y_train)
y_pred_ovr = ovr_model.predict(X_test)

# One-vs-One strategy
ovo_model = OneVsOneClassifier(SVC(kernel='linear'))
ovo_model.fit(X_train, y_train)
y_pred_ovo = ovo_model.predict(X_test)

# compare accuracies
print("One-vs-Rest Accuracy:", accuracy_score(y_test, y_pred_ovr))
print("One-vs-One Accuracy:", accuracy_score(y_test, y_pred_ovo))
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import log_loss

# load dataset
X, y = datasets.load_iris(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train Naive Bayes classifier
model = GaussianNB()
model.fit(X_train, y_train)

# predict probabilities
y_prob = model.predict_proba(X_test)

# evaluate using log loss
print("Log Loss:", log_loss(y_test, y_prob))
```

43- Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
X, y = datasets.load_iris(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train SVM classifier
model = SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# visualize confusion matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

44- Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute
Error (MAE) instead of MSE.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error

# load dataset
data = fetch_california_housing()
X, y = data.data, data.target

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train SVR model
model = SVR(kernel='rbf')
model.fit(X_train, y_train)

# make predictions
y_pred = model.predict(X_test)

# evaluate using MAE
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
```

45-Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC
score.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score

# load dataset
X, y = datasets.load_breast_cancer(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# predict probabilities
y_prob = model.predict_proba(X_test)[:, 1]

# evaluate using ROC-AUC score
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

46- Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# load dataset
X, y = datasets.load_breast_cancer(return_X_y=True)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# train SVM with probability estimates
model = SVC(kernel='linear', probability=True)
model.fit(X_train, y_train)

# get predicted probabilities
y_scores = model.predict_proba(X_test)[:, 1]

# calculate precision and recall
precision, recall, _ = precision_recall_curve(y_test, y_scores)

# plot precision-recall curve
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```
